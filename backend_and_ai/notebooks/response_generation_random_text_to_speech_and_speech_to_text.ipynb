```python
def random_response(input_question):
    responses={}
    responses.update(intent_responses_1)
    responses.update(intent_responses_2)
    if input_question in responses:
        chosen_set = random.choice([intent_responses_1, intent_responses_2])
        return chosen_set.get(input_question)
    else:
        return responses.get(input_question, "The requested information is not available.")
```

```python
def text_to_speech_response(response_text):
    tts = gTTS(response_text)
    audio_file = 'response.wav'
    tts.save(audio_file)
    return audio_file
```

```python
import speech_recognition as sr
```

```python
def speech_to_text():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Please speak now...")
        audio = r.listen(source)
    try:
        text = r.recognize_google(audio)
        print("You said: " + text)
        return text
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
        return ""
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
        return ""
```

```python
speech_text=speech_to_text()
if speech_text:
    response=random_response(speech_text)
    print(response)
```
